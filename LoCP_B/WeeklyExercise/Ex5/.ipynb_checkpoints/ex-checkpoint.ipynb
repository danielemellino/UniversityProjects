```python
def load_data_01(q, states,SPINS = True):
    fname = "DATA/dataRBM_q"+str(q)+"states"+str(states)+".csv"
    
    v = np.loadtxt(fname, delimiter = ",", dtype = int)
    v_original = np.loadtxt("DATA/dataRBM_original_states"+str(states)+".csv", delimiter = ",", dtype = int) # data before corruption

    N = len(v) #number of data
    L = len(v[0]) #length of samples
    print(f"each of N = {N} data samples has L = {L} digits")
    
    for n in range(10):
        print(v[n])
    print(".....")

    if SPINS:
        vmin = -1
        GAP = 2 #gap between the two categories
        v = 2*v - 1
        v_original = 2*v_original - 1
    else:
        vmin = 0
        GAP = 1

    #store initial values
    v0 = np.copy(v)

    for i in range(10):
        print(v[i])
    print(".....")
    
    return N, L, vmin, GAP, v_original, v0, v
```

```python
def create_coord(np, x0):
    x = [x0] * np
    y = list(range(np))
    for i in range(np):
        y[i] = y[i]/(np - 1.) - 0.5
    return(x, y)

# (x1, y1) = create_coord(L, 0) #visible units
# (x2, y2) = create_coord(M, 1) #hidden units

def mycolor(val):
    if val > 0:
        return "red" #red for positive weights
    elif val < 0:
        return "blue" #blue for negative weights
    return "black"

def plotgraph(w, a, b, epoch, M):
    
    A = 2. / w.max()
    for i in range(L): # visible units
        for j in range(M): # hidden units
            ex, ey, col = (x1[i], x2[j]), (y1[i], y2[j]), mycolor(w[i][j]) # coordinates and color
            plt.plot(ex, ey, col, zorder=1, lw=A*np.abs(w[i][j])) # lw proportional to the value of the weights
    
    A = 300./(a.max() + b.max())
    for i in range(L): # visible units
        plt.scatter(x1[i], y1[i], s = A * np.abs(a[i]), zorder = 2, c = mycolor(a[i])) # s = size proportional to values
    plt.title(f"> 0 red, < 0 blue, epoch = {epoch}")
    for j in range(M): # hidden units
        plt.scatter(x2[j], y2[j], s = A * np.abs(b[j]), zorder = 2, c = mycolor(b[j]))
    plt.title(f"> 0 red, < 0 blue, epoch = {epoch}")
    plt.show()
    return
```

```python
def activate(v_in, wei, bias, DE, beta = 1, info = False):
    act = np.dot(v_in, wei) + bias
    prob = 1. / (1 + np.exp(-DE*beta*act)) #probability to generate a 1
    n = len(act) # length of the activation
    v_out = np.full(n, vmin)
    v_out[np.random.random_sample(n) < prob] = 1
    if info:
        print("input =", v_in)
        print("act =", act)
        print("prob =", prob)
        print("output =", v_out)
    return (v_out)

# activate(v[0], w, b, GAP, info = True)
```

```python
def RBM_train(N, L, M, v, l_rate = 1.0, mini = 500, m = 0, nepoch = 50,STORE_w=False):

    np.random.seed(12345)
    
    # initialization
    sigma = np.sqrt(4./float(L + M))
    w = sigma * (2*np.random.rand(L, M) - 1)
    a = sigma * (2*np.random.rand(L) - 1)
    b = np.zeros(M)
    store_w=[] # per punto 5
    store_counter=[] #5
    
    
    np.random.seed(12345)
    
    for epoch in range(1, 1+nepoch):
        counter_m=0 # 5
        for n in range(N):
            if m == 0:
                # initialize
                #average of v according to the data and the model
                v_data, v_model = np.zeros(L), np.zeros(L)
                #average of h according to the data and the model
                h_data, h_model = np.zeros(M), np.zeros(M)
                #average of vh according to the data and the model
                vh_data, vh_model = np.zeros((L, M)), np.zeros((L, M))

            # positive CD phase
            h = activate(v[n], w, b, GAP)
            # negative CD phase
            vf = activate(h, w.T, a, GAP)
            # positive CD phase nr 2
            hf = activate(vf, w, b, GAP)

            v_data += v[n]
            v_model += vf
            h_data += h
            h_model += hf
            vh_data += np.outer(v[n].T, h)
            vh_model += np.outer(vf.T, hf)

            m += 1

            # gradient descent
            if m == mini:
                C = l_rate / mini
                dw = C*(vh_data - vh_model)
                da = C*(v_data - v_model)
                db = C*(h_data - h_model)

                if epoch <= 2 and n <= mini:
                    print("------ epoch =", epoch, "  n =", n, "  m =", m)
                    print("dw =", dw)
                    print("da =", da)
                    print("db =", db)
                if STORE_w==True :
                    store_w.append(dw)
                    counter_m=counter_m+1
                    store_counter.append([counter_m,epoch])
                
                    
                w += dw
                a += da
                b += db

                m = 0

        # randomize order
        np.random.shuffle(v)
        l_rate = l_rate / (0.05 * l_rate + 1)
        if epoch%5 == 0: 
            plotgraph(w, a, b, epoch, M)
            print("l_rate", l_rate)
        
    if STORE_w==True :
        return a,b,w,store_w,store_counter

            
    return a, b, w
```

```python
def performance(v0, a, b, w, phase, bet = 1):
    
    v1 = np.full((N, L), vmin)
    
    if phase == 0: # no lower beta
        score = 0
        for n in range(N):
            h = activate(v0[n], w, b, GAP)
            v1[n] = activate(h, w.T, a, GAP)
            if (v1[n] == v_original[n]).all():
#            if ((v1[n]==s1).all()) or ((v1[n]==s2).all()) or ((v1[n]==s3).all()) or ((v1[n]==s4).all()):
                score += 1
    
    elif phase == 1: # lower beta in negative phase
        score = 0
        for n in range(N):
            h = activate(v0[n], w, b, GAP)
            v1[n] = activate(h, w.T, a, GAP, beta = bet)
            if (v1[n] == v_original[n]).all():
#            if ((v1[n]==s1).all()) or ((v1[n]==s2).all()) or ((v1[n]==s3).all()) or ((v1[n]==s4).all()):
                score += 1
            
    elif phase == 2: # lower beta in both positive and negative phases
        score = 0
        for n in range(N):
            h = activate(v0[n], w, b, GAP, beta = bet)
            v1[n] = activate(h, w.T, a, GAP, beta = bet)
            if (v1[n] == v_original[n]).all():
#            if ((v1[n]==s1).all()) or ((v1[n]==s2).all()) or ((v1[n]==s3).all()) or ((v1[n]==s4).all()):
                score += 1 
    
    else: print("Error: phase can assume only value 0, 1 or 2")

    score = score/N
    print("Score =",score)
    print("--------------------------------------------------------")
    print("         Corrupted                 Fantasy")
    print("--------------------------------------------------------")
    for n in range(10):
        t = False
        if (v1[n] == v_original[n]).all():
        #if ((v1[n]==s1).all()) or ((v1[n]==s2).all()) or ((v1[n]==s3).all()) or ((v1[n]==s4).all()):
            t = True
        print((v0[n]).astype(int), (v1[n]).astype(int), t)

    return
```

```python
N, L, vmin, GAP, v_original, v0, v = load_data_01(q = 0.1, states = 4)

M = 3

(x1, y1) = create_coord(L, 0) #visible units
(x2, y2) = create_coord(M, 1) #hidden units

a, b, w = RBM_train(N, L, M, v)
```

### Beta = 1

```python
performance(v0, a, b, w, phase = 0)
```

### Case for lower temperature in negative phase

```python
performance(v0, a, b, w, phase = 1, bet = 2)
```

### Case for lower temperature in both negative and positive phases

```python
performance(v0, a, b, w, phase = 2, bet = 2)
```

### Deterministic case

```python
performance(v0, a, b, w, phase = 2, bet = np.inf)
```
